```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales_data.csv")   # replace with your dataset
df.head()
```

```python
print("Shape (rows, columns):", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nInfo:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
```

```python
product_sales = df.groupby("Product")["Sales"].sum()
region_sales = df.groupby("Region")["Sales"].sum()

print("Sales by Product:\n", product_sales)
print("\nSales by Region:\n", region_sales)
```

```python
high_sales = df[df["Sales"] > 1000]
high_sales
```

```python
# Bar chart
plt.figure(figsize=(8,5))
product_sales.plot(kind="bar", color="skyblue", title="Total Sales by Product")
plt.ylabel("Sales")
plt.xlabel("Product")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Pie chart
plt.figure(figsize=(6,6))
region_sales.plot(kind="pie", autopct="%1.1f%%", title="Sales by Region")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
